```python
class Solution:
    def strStr(self, haystack: str, needle: str) -> int:
        if not needle or haystack.__len__() < needle.__len__():
            return -1
        

        len_needle = len(needle)
        len_haystack = len(haystack)
        # print(needle[0:len_needle])
        # print(len_haystack)
        i = 0
        needle_str = needle[0:len_needle]
        j = len_needle

        while j != len_haystack + 1:
            if needle_str == haystack[i:j]:
                return i
            else:
                i = i + 1
                j = j + 1
        return -1
        # print(len_needle)
        # print(haystack[0:2])
        # print(needle[0:2])
        # print(haystack[2:4])
        # print(needle[:] == haystack[2:4])

s = Solution()
result = s.strStr("hello", "llogggggg")
print(result)  # Output: 2
```

Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

 

Example 1:

Input: haystack = "sadbutsad", needle = "sad"
Output: 0
Explanation: "sad" occurs at index 0 and 6.
The first occurrence is at index 0, so we return 0.
Example 2:

Input: haystack = "leetcode", needle = "leeto"
Output: -1
Explanation: "leeto" did not occur in "leetcode", so we return -1.
 

Constraints:

1 <= haystack.length, needle.length <= 104
haystack and needle consist of only lowercase English characters.
