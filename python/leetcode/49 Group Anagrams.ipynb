Given an array of strings strs, group the anagrams together. You can return the answer in any order.

 

Example 1:

Input: strs = ["eat","tea","tan","ate","nat","bat"]

Output: [["bat"],["nat","tan"],["ate","eat","tea"]]

Explanation:

There is no string in strs that can be rearranged to form "bat".
The strings "nat" and "tan" are anagrams as they can be rearranged to form each other.
The strings "ate", "eat", and "tea" are anagrams as they can be rearranged to form each other.
Example 2:

Input: strs = [""]

Output: [[""]]

Example 3:

Input: strs = ["a"]

Output: [["a"]]

 

Constraints:

1 <= strs.length <= 104
0 <= strs[i].length <= 100
strs[i] consists of lowercase English letters.



```python
from icecream import ic

from typing import List


class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        lenght = len(strs)
        result = []
        dict_str = {}

        while lenght != 0:
            popped_element_value = strs.pop()
            sorted_element_key = ''.join(sorted(popped_element_value))
            if sorted_element_key not in dict_str:  # Only initialize once
                dict_str[sorted_element_key] = []
            dict_str[sorted_element_key].append(popped_element_value)  # Add the first element
            lenght = len(strs)
            ic(dict_str)
            for str_in_strs in strs:
                temp = ''.join(sorted(str_in_strs))
                ic(temp)
                if temp == sorted_element_key:  # Compare sorted keys, not list with string
                        dict_str[sorted_element_key].append(str_in_strs)
                        ic(dict_str)
                        strs.remove(str_in_strs)
        for i in dict_str:
             result.append(dict_str[i])
        return result
        ic(result)
strs = ["eat","tea","tan","ate","nat","bat"]
solution = Solution()
result = solution.groupAnagrams(strs)
ic(result)
```
