Given the head of a sorted linked list, delete all duplicates such that each element appears only once. Return the linked list sorted as well.

 

Example 1:


Input: head = [1,1,2]
Output: [1,2]
Example 2:


Input: head = [1,1,2,3,3]
Output: [1,2,3]
 

Constraints:

The number of nodes in the list is in the range [0, 300].
-100 <= Node.val <= 100
The list is guaranteed to be sorted in ascending order.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class CreateList:
    def __init__(self):
        self.head = None
        self.size = 0

    def is_empty(self):
        return self.head is None
    
    def prepend(self, data): #(Insert at Head)
        new_node = ListNode(val=data)
        new_node.next = self.head
        self.head = new_node
        self.size += 1

    def append(self, data):
        new_node = ListNode(val=data)
        if self.is_empty():
            self.head = new_node
        else:
            current = self.head
            assert current is not None
            while current.next:
                current = current.next
            current.next = new_node
        self.size += 1


class Solution:
    def deleteDuplicates(self, head):
        # handle empty input (head expected as a Python list here)
        if not head:
            return []

        # build linked list from input list
        create_ll_1 = CreateList()
        for i in head:
            create_ll_1.append(i)
        
        # remove duplicates in-place
        current = create_ll_1.head
        while current is not None and current.next is not None:
            if current.val == current.next.val:
                # skip the duplicate node
                current.next = current.next.next
            else:
                current = current.next

        return create_ll_1.head





# head = [1,1,2]
head = [1,1,2,3,3,3,3]
solution = Solution()
result = solution.deleteDuplicates(head)


while result:
    print(result.val)
    result = result.next

# check the p1.val is == or < with p2.val - only proceed if == means - need to remove the duplicate.
```
