You are given the heads of two sorted linked lists list1 and list2.

Merge the two lists into one sorted list. The list should be made by splicing together the nodes of the first two lists.

Return the head of the merged linked list.

 

Example 1:


Input: list1 = [1,2,4], list2 = [1,3,4]
Output: [1,1,2,3,4,4]
Example 2:

Input: list1 = [], list2 = []
Output: []
Example 3:

Input: list1 = [], list2 = [0]
Output: [0]
 

Constraints:

The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
Both list1 and list2 are sorted in non-decreasing order.

````python
from typing import Optional

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class CreateList:
        def __init__(self):
            self.head = None
            self.size = 0
        def is_empty(self):
            return self.head is None
        def __len__(self):
            return self.size
        
        def prepend(self, data): #(Insert at Head)
            new_node = ListNode(val=data)
            new_node.next = self.head
            self.head = new_node
            self.size += 1

        def append(self, data):
            new_node = ListNode(val=data)
            if self.is_empty():
                self.head = new_node
            else:
                current = self.head
                assert current is not None
                while current.next:
                    current = current.next
                current.next = new_node
            self.size += 1

        def insert(self, index, data):
            if index < 0 or index > self.size:
                raise IndexError("Index out of bounds")
            
            if index == 0:
                self.prepend(data)
                return
            
            new_node = ListNode(data)
            current = self.head
            assert current is not None
            for i in range(index - 1):
                current = current.next
                assert current is not None
            
            new_node.next = current.next
            current.next = new_node
            self.size += 1

class Solution:
    def mergeTwoLists(self, list1, list2):
        def to_values(x):
            vals = []
            if x is None:
                return vals
            if isinstance(x, (list, tuple)):
                return list(x)
            # assume ListNode head
            cur = x
            while cur is not None:
                vals.append(cur.val)
                cur = cur.next
            return vals

        vals1 = to_values(list1)
        vals2 = to_values(list2)
        create_ll_1 = CreateList()
        create_ll_2 = CreateList()

        for i in vals1:
            create_ll_1.append(i)
        for i in vals2:
            create_ll_2.append(i)

        
        # print(create_ll_1.head.val == create_ll_2.head.val)
        # print(create_ll_1.head.next.val == create_ll_2.head.next.val)
        # print(create_ll_1.head.next.next.val == create_ll_2.head.next.next.val)
        merged_list = CreateList()
        ptr1 = create_ll_1.head
        ptr2 = create_ll_2.head
        while ptr1 is not None and ptr2 is not None:
            if ptr1.val >= ptr2.val:
                merged_list.append(ptr2.val)
                ptr2 = ptr2.next
            else:
                merged_list.append(ptr1.val)
                ptr1 = ptr1.next
        while ptr1 is not None:
            merged_list.append(ptr1.val)
            ptr1 = ptr1.next
        while ptr2 is not None:
            merged_list.append(ptr2.val)
            ptr2 = ptr2.next
        return merged_list.head

            

list1 = [1,2,4]
list2 = [1,3,4]
s = Solution()
result = s.mergeTwoLists(list1=list1, list2=list2)
 # Example to print the second value in the merged list
while result is not None:
    print(result.val)
    result = result.next
````

Here are **ASCII diagrams** that visualize **prepend** and **append** operations on a singly linked list. These sketches show exactly what changes in pointer structure.

---

## **1. Prepend (Insert at Head)**

You create a new node and make it point to the current head. Then you update `head` to this new node.

### **Before**

```
head
 |
 v
 A → B → C → None
```

### **Operation**

Create `X`, set:

```
X.next = head
head = X
```

### **After**

```
head
 |
 v
 X → A → B → C → None
```

The list grows from the **front**.

---

## **2. Append (Insert at Tail)**

You find the last node and connect it to the new node.

### **Before**

```
head
 |
 v
 A → B → C → None
```

### **Operation**

Create `X`, set:

```
C.next = X
X.next = None
```

### **After**

```
head
 |
 v
 A → B → C → X → None
```

The list grows at the **end**.

---

## Combined Perspective (Side-by-side)

```
PREPEND                          APPEND
--------                         --------
Before:                          Before:
A → B → C → None                 A → B → C → None

After prepend X:                 After append X:
X → A → B → C → None             A → B → C → X → None
```
