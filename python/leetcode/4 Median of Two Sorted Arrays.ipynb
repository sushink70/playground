4. Median of Two Sorted Arrays
Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.

The overall run time complexity should be O(log (m+n)).

 

Example 1:

Input: nums1 = [1,3], nums2 = [2]
Output: 2.00000
Explanation: merged array = [1,2,3] and median is 2.
Example 2:

Input: nums1 = [1,2], nums2 = [3,4]
Output: 2.50000
Explanation: merged array = [1,2,3,4] and median is (2 + 3) / 2 = 2.5.
 

Constraints:

nums1.length == m
nums2.length == n
0 <= m <= 1000
0 <= n <= 1000
1 <= m + n <= 2000
-106 <= nums1[i], nums2[i] <= 106

```python
from typing import List

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        arr = []
        lenght_1 = len(nums1) - 1
        lenght_2 = len(nums2) - 1
        i = 0
        j = 0
        lenght_3 = 9

        print(lenght_3)
        while True:
            if nums1[i] < nums2[j] and i < lenght_1:
                arr.append(nums1[i])
                i = i  + 1
            
            elif nums1[i] > nums2[j] and j < lenght_2:
                arr.append(nums2[j])
                j = j + 1
            else:
                break
            
        if lenght_1 < lenght_2:
            arr.extend(nums1[i:])
            arr.extend(nums2[j:])
        else:
            arr.extend(nums2[j:])
            arr.extend(nums1[i:])
        
        mid = len(arr) // 2
        print("mid:", mid)
        
        print(arr)

        print(arr[mid])
        print(arr[mid + 1])

        if mid % 2 == 0:
            return float(arr[mid] + arr[mid - 1] / 2)
        else:
            return float(arr[mid])


num2 = [1, 3, 5, 7, 9]
num1 = [2, 4, 6, 8, 10, 11, 12, 13, 14, 15, 16, 17, 18]
# num2 = [1, 3]
# num1 = [2]
# num1 = [1,2]
# num2 = [3,4]
soulution = Solution()
result = soulution.findMedianSortedArrays(num1, num2)
print(result)  # Expected output: 2.0
```
