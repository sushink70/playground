```python
import sys
import os

# Core path: Full executable path in the active env
print(f"Python Executable Path: {sys.executable}")

# Env root: Derived from the bin dir
env_root = os.path.dirname(os.path.dirname(sys.executable))
print(f"Virtual Environment Root: {env_root}")

# Bonus: Activated env var (None if not shell-activated, but venv still isolated)
print(f"VIRTUAL_ENV: {os.environ.get('VIRTUAL_ENV', 'Not set (but isolated via sys.path)')}")

# Kernel name for cross-reference
print(f"Kernel Display Name: {os.environ.get('JPY_PARENT_PID', 'N/A')}")
```
