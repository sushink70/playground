```python
class ArrayStack:

    def __init__(self, capacity=None):

        self._data = []
        self._capacity = capacity

    def push(self, item):
        if self._capacity is not None and len(self._data) >= self._capacity:
            raise OverflowError("Stack overflow: Cannot push to full stack")
        
        self._data.append(item)

    def pop(self):
        if self.is_empty():
            raise IndexError("Stack underflow: Cannot pop from empty stack")
        return self._data.pop()
    
    def peek(self):

        if self.is_empty():
            raise IndexError("Cannot peed empty stack")
        
        return self._data[-1]
    
    def is_empty(self):
        return len(self._data) == 0
    
    def size(self):
        return len(self._data)
    
    def clear(self):
        self._data.clear()

    def search(self, item):

        try:
            index = len(self._data) - 1 - self._data[::-1].index(item)
            return len(self._data) - index
        except ValueError:
            return -1
        
    def display(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            print("Stack (bottom to top):", self._data)

    def __str__(self):
        return f"Stack({self._data})"
    
    def __len__(self):
        return len(self._data)
    
    def __bool__(self):
        return not self.is_empty()
        
    
```
